```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import matplotlib
import numpy as np
```

```python
df=pd.read_excel("teste.xlsx",sheet_name="dia3")

tm=df['hr'][0:95] #manha ufes
tp=df['hr'][95:190] #manha pico

ti=df['hr'][190:285] #tarde ufes
tpi=df['tp'][190:285] #tarde ufes
ui=df['ur'][190:285]

tii=df['hr'][285:380] #tarde pico
tpii=df['tp'][285:380]
uii=df['ur'][285:380]

tei=df['hr'][380:475] #entardecer ufes
tpei=df['tp'][380:475]
uei=df['ur'][380:475]

teii=df['hr'][475:571] #entardecer pico
tpeii=df['tp'][475:571]
ueii=df['ur'][475:571]

tpm=df['tp'][0:95]
Tp=df['tp'][95:190]
tpe=df['tp'][190:]

um=df['ur'][0:95]
up=df['ur'][95:190]
ue=df['ur'][190:]
```

```python
startTime = df['hr'][0]
df['hr'] = pd.to_timedelta(df['hr'])
timeProgress = []
for timeStamp in df['hr']:
        timePassed = timeStamp - startTime
        timeProgress.append(timePassed)
custom_ticks = range(0, len(timeProgress), 15)
```

```python
fig, ax =plt.subplots(3,3, figsize =(12,7), constrained_layout=True)
ax[0,0].axes.set_xticks(custom_ticks)
ax[0,0].axes.set_yticks(np.arange(15, 50, 5))
ax[0,0].set_ylim(15,45)
ax1=ax[0,0].twinx()
ax[0,0].set_title('UFES $≈09:00$')
ax[0,0].tick_params(axis='x', rotation=45)
#ax[0,0].set_xlabel('Tempo')
ax[0,0].set_ylabel('°C', rotation=0, labelpad=12)
ax[0,0].plot(tm, tpm, color='red',linestyle='solid', label='$T_{MEM}$')
ax[0,0].scatter(47,20, marker='o',color='darkorange',label='$T_{A617}=20°C$',edgecolors='black')
ax[0,0].legend(bbox_to_anchor=(0.57,1.05),frameon=False,fontsize=11)
ax[0,0].tick_params(axis='y')
ax[0,0].grid(axis='y')
ax1.set_ylim(40,100)
ax1.axes.set_yticks(np.arange(40, 110, 
                              10))
ax1.plot(tm, um, color='blue',linestyle='solid',label='$UR_{MEM}$')
ax1.set_ylabel('%', rotation=0)
ax1.scatter(47,72, marker='o',color='cyan', label='$UR_{A617}=72\%$',edgecolors='black')
ax1.legend(loc=3,bbox_to_anchor=(0.48,0.58),frameon=False,fontsize=11)
labels=['09:00', '10:00']
x = np.arange(len(labels))
width = 0.35
terr=[5,5.99]
uerr=[13.88, 3.44]
rects1=ax[0,2].bar(x - width/2,terr, width, label='Temperatura', color='red',edgecolor = 'black')
rects2=ax[0,2].bar(x + width/2, uerr, width, label='Umidade', color='blue',edgecolor = 'black')
ax[0,2].set_ylabel('Porcentagem', labelpad=-4)
ax[0,2].set_title('Erro(%) - Etapa 1')
ax[0,2].set_xticks(x, labels)
ax[0,2].legend()
ax[0,2].set_ylim(0,100)
ax[0,2].bar_label(rects1,fmt='%.1f%%', padding=3, fontsize=11)
ax[0,2].bar_label(rects2,fmt='%.1f%%', padding=3, fontsize=11)

ax[0,1].axes.set_xticks(custom_ticks)
ax[0,1].set_ylim(15,45)
ax[0,1].axes.set_yticks(np.arange(15, 50, 5))
ax[0,1].set_title('P7 $≈ 10:00$')
ax[0,1].tick_params(axis='x', rotation=45)
ax2=ax[0,1].twinx()
ax[0,1].set_ylabel('°C', rotation=0, labelpad=12)
ax[0,1].plot(tp, Tp, color='red',linestyle='solid', label='$T_{MEM}$')
ax[0,1].scatter(28,21.7, marker='o',color='orange',label='$T_{A617}=21,7°C$',edgecolor = 'black')
ax[0,1].legend(bbox_to_anchor=(0.57,1.05),frameon=False,fontsize=11)
ax[0,1].tick_params(axis='y')
ax[0,1].grid(axis='y')
ax2.set_ylim(30, 90)
ax2.axes.set_yticks(np.arange(30, 100, 10))
ax2.plot(tp, up, color='blue',linestyle='solid',label='$UR_{MEM}$')
ax2.set_ylabel('%', rotation=0)
ax2.scatter(28,58, marker='o',color='cyan', label='$UR_{A617}=58\%$',edgecolor = 'black')
ax2.legend(loc=3,bbox_to_anchor=(0.48,0.58), frameon=False,fontsize=11)

ax[1,0].axes.set_xticks(custom_ticks)
ax[1,0].set_ylim(15,45)
ax[1,0].axes.set_yticks(np.arange(15, 50, 5))
ax[1,0].set_title('UFES $≈ 13:00$')
ax[1,0].tick_params(axis='x', rotation=45)
ax2=ax[1,0].twinx()
ax[1,0].set_ylabel('°C', rotation=0, labelpad=12)
ax[1,0].plot(ti, tpi, color='red',linestyle='solid', label='$T_{MEM}$')
ax[1,0].scatter(26,24.9, marker='o',color='orange',label='$T_{A617}=24.9°C$',edgecolor = 'black')
ax[1,0].legend(bbox_to_anchor=(0.57,1.05),frameon=False,fontsize=11)
ax[1,0].tick_params(axis='y')
ax[1,0].grid(axis='y')
ax2.set_ylim(30, 90)
ax2.axes.set_yticks(np.arange(30, 100, 10))
ax2.plot(ti, ui, color='blue',linestyle='solid',label='$UR_{MEM}$')
ax2.set_ylabel('%', rotation=0)
ax2.scatter(26,39, marker='o',color='cyan', label='$UR_{A617}=39\%$',edgecolor = 'black')
ax2.legend(loc=3,bbox_to_anchor=(0.48,0.58), frameon=False,fontsize=11)



ax[1,1].axes.set_xticks(custom_ticks)
ax[1,1].set_ylim(15,45)
ax[1,1].axes.set_yticks(np.arange(15, 50, 5))
ax4=ax[1,1].twinx()
ax[1,1].set_title('P7 $≈14:00$')
ax[1,1].tick_params(axis='x', rotation=45)
#ax[0,0].set_xlabel('Tempo')
ax[1,1].set_ylabel('°C', rotation=0, labelpad=12)
ax[1,1].plot(tii, tpii, color='red',linestyle='solid', label='$T_{MEM}$')
ax[1,1].scatter(19,25.8, marker='o',color='orange',label='$T_{A617}=25,8°C$',edgecolors='black')
ax[1,1].legend(bbox_to_anchor=(0.56,1.06),frameon=False,fontsize=11)
ax[1,1].tick_params(axis='y')
ax[1,1].grid(axis='y')
ax4.set_ylim(30, 90)
ax4.axes.set_yticks(np.arange(30, 100, 10))
ax4.plot(tii, uii, color='blue',linestyle='solid',label='$UR_{MEM}$')
ax4.set_ylabel('%', rotation=0)
ax4.scatter(19,34, marker='o',color='cyan', label='$UR_{A617}=34\%$',edgecolors='black')
ax4.legend(loc=3,bbox_to_anchor=(0.46,0.58),frameon=False,fontsize=11)


#Bar Chart
labels=['13:00', '14:00']
x = np.arange(len(labels))
width = 0.35
terr=[8.43, 4.65]
uerr=[5.12, 17.94]
rects1=ax[1,2].bar(x - width/2,terr, width, label='Temperatura', color='red',edgecolor = 'black')
rects2=ax[1,2].bar(x + width/2, uerr, width, label='Umidade', color='blue',edgecolor = 'black')
ax[1,2].set_ylabel('Porcentagem', labelpad=-4)
ax[1,2].set_title('Erro(%) - Etapa 2')
ax[1,2].set_xticks(x, labels)
ax[1,2].legend()
ax[1,2].set_ylim(0,100)
ax[1,2].bar_label(rects1,fmt='%.1f%%', padding=3,fontsize=11)
ax[1,2].bar_label(rects2,fmt='%.1f%%', padding=3,fontsize=11)

ax[2,0].axes.set_xticks(custom_ticks)
ax[2,0].set_ylim(15,45)
ax[2,0].axes.set_yticks(np.arange(15, 50, 5))
ax2=ax[2,0].twinx()
ax[2,0].set_title('UFES $≈16:00$')
ax[2,0].tick_params(axis='x', rotation=45)
ax[2,0].set_ylabel('°C', rotation=0, labelpad=12)
ax[2,0].plot(tei, tpei, color='red',linestyle='solid', label='$T_{MEM}$')
ax[2,0].scatter(37,25.3, marker='o',color='orange',label='$T_{A617}=25,3C$',edgecolors='black')
ax[2,0].legend(bbox_to_anchor=(0.55,1.06),frameon=False,fontsize=11)
ax[2,0].tick_params(axis='y')
ax[2,0].grid(axis='y')
ax2.set_ylim(30, 90)
ax2.axes.set_yticks(np.arange(30, 100, 10))
ax2.plot(tei, uei, color='blue',linestyle='solid',label='$UR_{MEM}$')
ax2.set_ylabel('%', rotation=0)
ax2.scatter(37,39, marker='o',color='cyan', label='$UR_{A617}=39\%$',edgecolors='black')
ax2.legend(loc=3,bbox_to_anchor=(0.46,0.58),frameon=False,fontsize=11)

ax[2,1].axes.set_xticks(custom_ticks)
ax[2,1].set_ylim(15,45)
ax[2,1].axes.set_yticks(np.arange(15, 50, 5))
ax2=ax[2,1].twinx()
ax[2,1].set_title('P7 $≈17:00$')
ax[2,1].tick_params(axis='x', rotation=45)
#ax[0,0].set_xlabel('Tempo')
ax[2,1].set_ylabel('°C', rotation=0, labelpad=12)
ax[2,1].plot(teii, tpeii, color='red',linestyle='solid', label='$T_{MEM}$')
ax[2,1].scatter(30,20.2, marker='o',color='orange',label='$T_{A617}=20,2C$',edgecolor = 'black')
ax[2,1].legend(bbox_to_anchor=(0.55,1.06),frameon=False,fontsize=11)
ax[2,1].tick_params(axis='y')
ax[2,1].grid(axis='y')
ax2.set_ylim(30, 90)
ax2.axes.set_yticks(np.arange(30, 100, 10))
ax2.plot(teii, ueii, color='blue',linestyle='solid',label='$UR_{MEM}$')
ax2.set_ylabel('%', rotation=0)
ax2.scatter(30,64, marker='o',color='cyan', label='$UR_{A617}=64\%$',edgecolor = 'black')
ax2.legend(loc=3,bbox_to_anchor=(0.46,0.58),frameon=False,fontsize=11)


#Bar Chart 
labels=['16:00', '17:00']
x = np.arange(len(labels))
width = 0.35
terr=[13,3.9]
uerr=[20.51, 25]
rects1=ax[2,2].bar(x - width/2,terr, width, label='Temperatura', color='red',edgecolor = 'black')
rects2=ax[2,2].bar(x + width/2, uerr, width, label='Umidade', color='blue',edgecolor = 'black')
ax[2,2].set_ylabel('Porcentagem', labelpad=-4)
ax[2,2].set_title('Erro(%) - Etapa 3')
ax[2,2].set_xticks(x, labels)
ax[2,2].legend()
ax[2,2].set_ylim(0,100)
ax[2,2].bar_label(rects1,fmt='%.1f%%', padding=3,fontsize=11)
ax[2,2].bar_label(rects2,fmt='%.1f%%', padding=3,fontsize=11)
plt.savefig('0707.png',facecolor='w',dpi=300)
plt.show()
```
