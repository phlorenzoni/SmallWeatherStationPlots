```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gs
import datetime as dt
import matplotlib
import numpy as np
df=pd.read_excel("teste.xlsx",sheet_name="dia1")

tm=df['hr'][0:95] #manha ufes
tp=df['hr'][95:190] #manha pico

ti=df['hr'][190:285] #tarde ufes
tpi=df['tp'][190:285] #tarde ufes
ui=df['ur'][190:285]

tii=df['hr'][285:380] #tarde pico
tpii=df['tp'][285:380]
uii=df['ur'][285:380]

tei=df['hr'][380:475] #entardecer ufes
tpei=df['tp'][380:475]
uei=df['ur'][380:475]

teii=df['hr'][475:571] #entardecer pico
tpeii=df['tp'][475:571]
ueii=df['ur'][475:571]

tpm=df['tp'][0:95]
Tp=df['tp'][95:190]
tpe=df['tp'][190:]

um=df['ur'][0:95]
up=df['ur'][95:190]
ue=df['ur'][190:]
startTime = df['hr'][0]
df['hr'] = pd.to_timedelta(df['hr'])
timeProgress = []
for timeStamp in df['hr']:
        timePassed = timeStamp - startTime
        timeProgress.append(timePassed)
custom_ticks = range(0, len(timeProgress), 15)
```
