```python
uf1 = {'Etapa 1':lx , 'Etapa 2':lxi,'Etapa 3':lxei}

fig, ax =plt.subplots(2,3, figsize =(8,6), constrained_layout=True)
#plt.suptitle('Variabilidade de Ilumin√¢ncia e UV na UFES')
ax[0,0].set_title('05/07/21')
ax[0,0].boxplot(uf1.values(),showfliers=False)
ax[0,0].set_xticklabels(uf1.keys())
ax[0,0].set_ylim(930,1000)
ax[0,0].set_ylabel('Lux', rotation=0, labelpad=12)

uf2={'Etapa 1':lxm , 'Etapa 2':lxt,'Etapa 3':lxe}
ax[0,1].set_title('06/07/21')
ax[0,1].boxplot(uf2.values(),showfliers=False)
ax[0,1].set_xticklabels(uf2.keys())
ax[0,1].set_ylim(700,1000)
ax[0,1].set_ylabel('Lux', rotation=0, labelpad=12)

uf3={'Etapa 1':lxm3 , 'Etapa 2':lxt3,'Etapa 3':lxe3}
ax[0,2].set_title('07/07/21')
ax[0,2].boxplot(uf3.values(),showfliers=False)
ax[0,2].set_xticklabels(uf3.keys())
ax[0,2].set_ylabel('Lux', rotation=0, labelpad=12)
ax[0,2].set_ylim(950,1000)

uf11 = {'Etapa 1':mv , 'Etapa 2':uvi,'Etapa 3':uvei}
ax[1,0].set_title('05/07/21')
ax[1,0].boxplot(uf11.values(),showfliers=False)
ax[1,0].set_xticklabels(uf11.keys())
ax[1,0].set_ylim(0,160)
ax[1,0].set_ylabel('mV', rotation=0,labelpad=12)

uf22 = {'Etapa 1':uvm , 'Etapa 2':uvt}
ax[1,1].set_title('06/07/21')
ax[1,1].boxplot(uf22.values(),showfliers=False)
ax[1,1].set_xticklabels(uf22.keys())
ax[1,1].set_ylim(0,120)
uf33 = {'Etapa 1':uvm3 , 'Etapa 2':uvt3}
ax[1,1].set_ylabel('mV', rotation=0,labelpad=10)

ax[1,2].boxplot(uf33.values(),showfliers=False)
ax[1,2].set_title('07/07/21')
ax[1,2].set_xticklabels(uf33.keys())
ax[1,2].set_ylim(40,140)
ax[1,2].set_ylabel('mV', rotation=0,labelpad=10)
#plt.savefig('BPUFES.PNG',dpi=450, facecolor='w')
plt.show()
```
